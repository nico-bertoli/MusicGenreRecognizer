```python
import os
import numpy
import librosa
import librosa.display
import IPython.display as ipd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
from tensorflow import keras
```

```python
    NDIGITS = 4               #digits of the number of files (if I have 2000 files it's 4)
    N_INPUT_FILES = 6000      #    (numero file del test set) 
    GENERE='pp'



    INPUT_FOLDER = '1wav15/'+GENERE          #name of the folder that contains wav files
    FILEBEGINNING = GENERE+'_'                #beginnig of the name of each file
    OUTPUT_FOLDER = '2timestamps/'+GENERE+'/'
```

```python
#TODO tradurre in inglese da qui in poi
#FUNCTIONS
#---------------------------------------------------------------------------------------GET ADDRESSES
# returns a lit with the names of wav files
def getAddresses (folderName,fileName,nFile = N_INPUT_FILES,nDigits = NDIGITS):
    #str(x).zfill adds zeros at the beginning of the number (es 4 -> 0004)
    n = [str(x).zfill(nDigits) for x in range(1,nFile+1)]

    #list with name of files
    ris = []
    for i in n:
       ris.append(f'{folderName}/{fileName}{i}.wav')
    return ris

#---------------------------------------------------------------------------------------GET AUDIOS
# input:
# adresses -> list with the names of wav files
#
# output:
# - wav files timestamps
# - theyr sample rate (I assume they have all the same sr)
def getAudios(addresses,nfiles = N_INPUT_FILES):
    ris = []
    #returns the timestamp and sample rate of the file
    inutile,sr = librosa.load(addresses[0]) 

    #TODO: PROVA A RIMUOVORE inutile
    for x in addresses:
        audio,inutile = librosa.load(x)
        ris.append(audio)
        print(f'{nfiles} files remaining')
        nfiles-=1

    return ris,sr
```

```python
#GET THE ADRESSES OF ALL FILES (A LIST FOR EACH GENRE)
addresses = getAddresses(INPUT_FOLDER,FILEBEGINNING)

#formato ottenuto:
print(f'tipo del formato ottenuto: {addresses[0]}')
```

```python
if len(addresses)==N_INPUT_FILES : print('addresses calculated with success')
```

```python
#GET THE AUDIOS
audios,sr = getAudios(addresses)

#ottengo un array di audio time series
#un audio time serie e' un array di valori che corrispondono ai sample del file nel tempo
```

```python
#==========================================
#   CALCOLO NUOVA NUMERAZIONE RANDOMICA   #
#==========================================
```

```python
gia_inseriti = []
for i in range (N_INPUT_FILES):
    gia_inseriti.append(False)
print(f'gia_inseriti ha: {len(gia_inseriti)} elementi')
print(gia_inseriti)
```

```python
numeri_randomici = []
for i in range(N_INPUT_FILES):numeri_randomici.append(0)
print(f'numeri_randomici ha {len(numeri_randomici)} elementi')
print(numeri_randomici)
```

```python
for n_file in range (N_INPUT_FILES):
    while True:
        random = np.random.randint(N_INPUT_FILES)  #crea un numero randomico tra 0 ed NFILES -1
        if gia_inseriti[random] == False:
            gia_inseriti[random] = True
            numeri_randomici[n_file]=random+1
            print(f'aggiungo:{random}in posizione:{n_file}')
            break
print(numeri_randomici)
```

```python
#==========================================
#               SALVO I FILE              #
#==========================================
```

```python
#salvataggio timestamps
audiosNump = numpy.array(audios[0:N_INPUT_FILES])

print(f'sample rate:{sr}')
print(f'ho {audiosNump.shape[0]} file lunghi {audiosNump.shape[1]}\n\n\n')
print(audiosNump)
```

```python
for n_file in range (N_INPUT_FILES):
    out_location = OUTPUT_FOLDER+str(numeri_randomici[n_file])+'.npy' 
    np.save(out_location,audiosNump[n_file])
```
